```python
def make_contours(fit, datasets, result, npoints, sigmas):
    cts_sigma = []
    for sigma in sigmas:
        contours = dict()
        for par_1, par_2 in combinations(["amplitude", "index", "lambda_"], r=2):
            idx1, idx2 = datasets.models.parameters.index(par_1), datasets.models.parameters.index(
                par_2
            )
            name1 = datasets.models.parameters_unique_names[idx1]
            name2 = datasets.models.parameters_unique_names[idx2]
            contour = fit.stat_contour(
                datasets=datasets,
                x=datasets.models.parameters[par_1],
                y=datasets.models.parameters[par_2],
                numpoints=npoints,
                sigma=sigma,
            )
            contours[f"contour_{par_1}_{par_2}"] = {
                par_1: contour[name1].tolist(),
                par_2: contour[name2].tolist(),
            }
        cts_sigma.append(contours)
    return cts_sigma
```

```python
%%time
from itertools import combinations
sigmas = [1, 2]
cts_sigma_array = []

for d in datasets_array:
    fit = Fit(store_trace=False)
    minuit_opts = {"tol": 0.001, "strategy": 2}
    fit.backend = "minuit"
    fit.optimize_opts = minuit_opts
    result = fit.run([d])
    print(result.success)
    cts_sigma = make_contours(fit=fit, datasets=d, result=result, npoints=11, sigmas=sigmas)
    cts_sigma_array.append(cts_sigma)
```

```python
pars = {
    "amplitude": r"$\phi_0 \,/\,(10^{-11}\,{\rm TeV}^{-1} \, {\rm cm}^{-2} {\rm s}^{-1})$",
    "index": r"$\Gamma",
    "lambda_": r"$\lambda$",
}

panels = [
    {
        "x": "amplitude",
        "y": "index",
        "cx": (lambda ct: ct["contour_index_amplitude"]["index"]),
        "cy": (lambda ct: np.array(1e11) * ct["contour_index_amplitude"]["amplitude"]),
    },
    {
        "x": "lambda_",
        "y": "phi",
        "cx": (lambda ct: ct["contour_lambda__amplitude"]["lambda_"]),
        "cy": (lambda ct: np.array(1e11) * ct["contour_lambda__amplitude"]["amplitude"]),
    },
    {
        "x": "index",
        "y": "lambda_",
        "cx": (lambda ct: ct["contour_index_lambda_"]["index"]),
        "cy": (lambda ct: ct["contour_index_lambda_"]["beta"]),
    },
]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
colors = ["m", "b", "c"]
for p, ax in zip(panels, axes):
    xlabel = pars[p["x"]]
    ylabel = pars[p["y"]]
    for i, d in enumerate(datasets):
        cts_sigma = cts_sigma_array[i]
        for ks in range(len(cts_sigma)):
            plot_contour_line(
                ax,
                p["cx"](cts_sigma[ks]),
                p["cy"](cts_sigma[ks]),
                lw=2.5,
                color=colors[ks],
                label=f"{d}:" + f"{sigmas[ks]}" + r"$\sigma$",
            )
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
plt.legend()
fig.savefig("plots/2a_contours_all")
plt.tight_layout()
```
