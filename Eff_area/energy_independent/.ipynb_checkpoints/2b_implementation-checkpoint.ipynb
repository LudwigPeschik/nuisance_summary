```python
def create_dataset_N_method(dataset_asimov, factor_neg, factor_pos, rnd, sigma_i, sigma_a, sigma_e, nuisance_full):
    sys_d_cor = sys_dataset(dataset_asimov, 0, rnd)
    dataset_cor_N = sys_d_cor.create_dataset_N()

    sys_d_neg = sys_dataset(dataset_asimov, factor_neg, rnd)
    dataset_neg_N = sys_d_neg.create_dataset_N()

    sys_d_pos = sys_dataset(dataset_asimov, factor_pos, rnd)
    dataset_pos_N = sys_d_pos.create_dataset_N()

    if nuisance_full == 1:
        dataset_cor_N.models.parameters['index_nuisance'].frozen = True
        dataset_cor_N.models.parameters['E_nuisance'].frozen = True
        dataset_neg_N.models.parameters['index_nuisance'].frozen = True
        dataset_neg_N.models.parameters['E_nuisance'].frozen = True
        dataset_pos_N.models.parameters['index_nuisance'].frozen = True
        dataset_pos_N.models.parameters['E_nuisance'].frozen = True
    elif nuisance_full == 2:
        dataset_cor_N.models.parameters['E_nuisance'].frozen = True
        dataset_neg_N.models.parameters['E_nuisance'].frozen = True
        dataset_pos_N.models.parameters['E_nuisance'].frozen = True
    

    penalising_invcovmatrix = [[1/sigma_i**2, 0, 0], [0, 1/sigma_a**2, 0], [0, 0, 1/sigma_e**2]]

    fig, ax = plt.subplots(1,1)
    #ax = fig.add_subplot(111)
    #cax = ax[0].matshow(penalising_covmatrix)  # interpolation='nearest')
    #fig.colorbar(cax);
    cax = ax.matshow(penalising_invcovmatrix)  # interpolation='nearest')
    fig.colorbar(cax);

    dataset_pos_N.penalising_invcovmatrix = penalising_invcovmatrix
    dataset_cor_N.penalising_invcovmatrix = penalising_invcovmatrix
    dataset_neg_N.penalising_invcovmatrix = penalising_invcovmatrix
    
    print("corr")
    fit_cor = Fit(store_trace=False)
    result_cor = fit_cor.run(dataset_cor_N)

    print("neg")
    fit_neg = Fit(store_trace=False)
    result_neg = fit_neg.run(dataset_neg_N)

    print("pos")
    fit_pos = Fit(store_trace=False)
    result_pos = fit_pos.run(dataset_pos_N)

    dataset_cor_N.models.covariance.get_subcovariance(dataset_cor_N.models.parameters.free_parameters).plot_correlation()
    dataset_neg_N.models.covariance.get_subcovariance(dataset_neg_N.models.parameters.free_parameters).plot_correlation()
    dataset_pos_N.models.covariance.get_subcovariance(dataset_pos_N.models.parameters.free_parameters).plot_correlation()
    
    datasets  = [dataset_pos_N, dataset_cor_N, dataset_neg_N]
    
    return datasets
```

```python
nuisance_full = 0
datasets = create_dataset_N_method(dataset_asimov, factor_neg, factor_pos, False, sigma, sigma, sigma, nuisance_full)
dataset_pos_N = datasets[0]
dataset_cor_N = datasets[1]
dataset_neg_N = datasets[2]
```
