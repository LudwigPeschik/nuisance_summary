```python

valuies = [norms, amplitudes, indices, lambdas]

valuies_standard = [[dataset_standard.models[3].parameters['norm'].value, 
                   dataset_standard.models[3].parameters['norm'].error],
                  [dataset_standard.models[0].parameters['amplitude'].value, 
                   dataset_standard.models[0].parameters['amplitude'].error],
                  [dataset_standard.models[0].parameters['index'].value, 
                   dataset_standard.models[0].parameters['index'].error],
                  [dataset_standard.models[0].parameters['lambda_'].value, 
                   dataset_standard.models[0].parameters['lambda_'].error]]


valuies_standard_N = [[dataset_standard_N.models[3].parameters['norm'].value, 
                   dataset_standard_N.models[3].parameters['norm'].error],
                  [dataset_standard_N.models[0].parameters['amplitude'].value, 
                   dataset_standard_N.models[0].parameters['amplitude'].error],
                  [dataset_standard_N.models[0].parameters['index'].value, 
                   dataset_standard_N.models[0].parameters['index'].error],
                  [dataset_standard_N.models[0].parameters['lambda_'].value, 
                   dataset_standard_N.models[0].parameters['lambda_'].error]]

is_within_norm = []
is_within_amplitude = []
is_within_index = []
is_within_lambda = []
is_within_norm_N = []
is_within_amplitude_N = []
is_within_index_N = []
is_within_lambda_N = []


i = 0
for n in norms:
    larger = np.all(n[0] >= (valuies_standard[0][0] - valuies_standard[0][1]))
    smaller = np.all(n[0] <= (valuies_standard[0][0] + valuies_standard[0][1]))
    is_within_norm.append(larger and smaller)
    
    larger = np.all(n[0] >= (valuies_standard_N[0][0] - valuies_standard_N[0][1]))
    smaller = np.all(n[0] <= (valuies_standard_N[0][0] + valuies_standard_N[0][1]))
    is_within_norm_N.append(larger and smaller)
    
for a in amplitudes:
    larger = np.all(a[0] >= (valuies_standard[1][0] - valuies_standard[1][1]))
    smaller = np.all(a[0] <= (valuies_standard[1][0] + valuies_standard[1][1]))
    is_within_amplitude.append(larger and smaller)
    
    larger = np.all(a[0] >= (valuies_standard_N[1][0] - valuies_standard_N[1][1]))
    smaller = np.all(a[0] <= (valuies_standard_N[1][0] + valuies_standard_N[1][1]))
    is_within_amplitude_N.append(larger and smaller)
    
for i in indices:
    larger = np.all(i[0] >= (valuies_standard[2][0] - valuies_standard[2][1]))
    smaller = np.all(i[0] <= (valuies_standard[2][0] + valuies_standard[2][1]))
    is_within_index.append(larger and smaller)
    
    larger = np.all(i[0] >= (valuies_standard_N[2][0] - valuies_standard_N[2][1]))
    smaller = np.all(i[0] <= (valuies_standard_N[2][0] + valuies_standard_N[2][1]))
    is_within_index_N.append(larger and smaller)
    
for l in lambdas:
    larger = np.all(l[0] >= (valuies_standard[3][0] - valuies_standard[3][1]))
    smaller = np.all(l[0] <= (valuies_standard[3][0] + valuies_standard[3][1]))
    is_within_lambda.append(larger and smaller)
    
    larger = np.all(l[0] >= (valuies_standard_N[3][0] - valuies_standard_N[3][1]))
    smaller = np.all(l[0] <= (valuies_standard_N[3][0] + valuies_standard_N[3][1]))
    is_within_lambda_N.append(larger and smaller)
    
within_norm = [np.count_nonzero(is_within_norm)/N,
              np.count_nonzero(is_within_norm_N)/N]

within_amplitude = [np.count_nonzero(is_within_amplitude)/N,
              np.count_nonzero(is_within_amplitude_N)/N]

within_index = [np.count_nonzero(is_within_index)/N,
              np.count_nonzero(is_within_index_N )/N]

within_lambda = [np.count_nonzero(is_within_lambda)/N,
              np.count_nonzero(is_within_lambda_N )/N]

withins = [within_norm, 
          within_amplitude,
          within_index,
          within_lambda]
```
