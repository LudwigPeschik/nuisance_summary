```python
with open(outputfolder+outputfile_gaussian, 'r') as f:
    data_gaussian = json.load(f)
amplitudes_gaussian = list(data_gaussian.keys())  
amplitudes_x_gaussian = [(float(a) * 1e12) for a in data_gaussian.keys()]

with open(outputfolder+outputfile_point, 'r') as f:
    data_point = json.load(f)
amplitudes_point = list(data_point.keys())  
amplitudes_x_point = [(float(a) * 1e12) for a in data_point.keys()]

with open(outputfolder+outputfile_offgaussian, 'r') as f:
    data_offgaussian = json.load(f)
amplitudes_offgaussian = list(data_offgaussian.keys())  
amplitudes_x_offgaussian = [(float(a) * 1e12) for a in data_offgaussian.keys()]

with open(outputfolder+outputfile_largegaussian, 'r') as f:
    data_largegaussian = json.load(f)
amplitudes_largegaussian = list(data_largegaussian.keys())  
amplitudes_x_largegaussian = [(float(a) * 1e12) for a in data_largegaussian.keys()]

def get_mask(spatial_model_type):
    dataset_asimov_testing = dataset_N_import.copy() 
    geom = dataset_asimov_testing.geoms['geom']
    center_coord = geom.center_skydir
    if spatial_model_type == "pointsource_center":
        skyregion = CircleSkyRegion(center=center_coord, radius = 0.3 *u.deg)
    if spatial_model_type ==  "CompleteFoV":
        skyregion = CircleSkyRegion(center=center_coord, radius = 20.3 *u.deg)
    if spatial_model_type == "gaussiansource_center":
        skyregion = CircleSkyRegion(center=center_coord, radius = 0.5 *u.deg)
    if spatial_model_type == "gaussiansource_offcenter":
        center_coord = SkyCoord(83.6*u.deg + 1*u.deg, 
                                22.0*u.deg + 1*u.deg,
                                frame = "icrs")
        skyregion = CircleSkyRegion(center=center_coord, radius = 0.5 *u.deg)
    if spatial_model_type == "largegaussiansource_center":
        skyregion = CircleSkyRegion(center=center_coord, radius = 2.0 *u.deg)
        
        

    map_mask= Map.from_geom(geom, 
                                data=geom.region_mask([skyregion], inside=True).data)

    return map_mask


def compute_signal_to_noise(amplitude  ,spatial_model_type , bg_norm = 1, index =2.4):
    if spatial_model_type == "pointsource_center":
        spatial_model_asimov = PointSpatialModel(lon_0=83.6*u.deg, 
                                                 lat_0=22.0*u.deg, 
                                                 frame='icrs')

    if spatial_model_type == "gaussiansource_center":
        spatial_model_asimov = GaussianSpatialModel(lon_0=83.6*u.deg, 
                                                 lat_0=22.0*u.deg, 
                                                sigma = 0.5 *u.deg,
                                                 frame='icrs')

    if spatial_model_type == "gaussiansource_offcenter":
        spatial_model_asimov = GaussianSpatialModel(lon_0=83.6*u.deg + 1*u.deg, 
                                             lat_0=22.0*u.deg + 1*u.deg, 
                                            sigma = 0.5 *u.deg,
                                             frame='icrs')
        
    if spatial_model_type == "largegaussiansource_center":
        spatial_model_asimov = GaussianSpatialModel(lon_0=83.6*u.deg, 
                                                 lat_0=22.0*u.deg, 
                                                sigma = 2 *u.deg,
                                                 frame='icrs')    
    if spatial_model_type != 'CompleteFoV':
        spectral_model_asimov = PowerLawSpectralModel(index=index,
                                                      amplitude=amplitude, 
                                                      reference="1 TeV") 
        model_asimov_testing = SkyModel(spatial_model=spatial_model_asimov, spectral_model=spectral_model_asimov, name="source")
        models_testing = Models(model_asimov_testing) 
    else:
        models_testing = Models() 



    dataset_asimov_testing = dataset_N_import.copy() 

    dataset_asimov_testing = MapDatasetNuisance(
        background=dataset_N_import.background,
        exposure=dataset_N_import.exposure,
        psf=dataset_N_import.psf,
        edisp=dataset_N_import.edisp,
        mask_fit=dataset_N_import.mask_fit,
        mask_safe=dataset_N_import.mask_safe,
        inv_corr_matrix=dataset_N_import.inv_corr_matrix,
        nuisance_mask=dataset_N_import.nuisance_mask,
    )


    bkg_model = FoVBackgroundModel(dataset_name=dataset_asimov_testing.name)
    bkg_model.parameters['tilt'].frozen =False

    models_testing.append(bkg_model)
    dataset_asimov_testing.models = models_testing

    ## setting the counts to npred = bg(norm) + sys + new source
    sys_ = (dataset_N_import.background - dataset_N_import.npred_background())
    dataset_asimov_testing.counts = dataset_N_import.background* bg_norm + dataset_asimov_testing.npred_signal() +sys_


    ## setting the nuisance parameters to zero 
    N_pars_zero = Parameters( [Parameter(name = par.name, value =0,frozen = par.frozen)  for par in dataset_N_import.N_parameters])
    dataset_asimov_testing.N_parameters = N_pars_zero
    
    map_mask = get_mask(spatial_model_type)
    bg = (dataset_asimov_testing.background*bg_norm*map_mask).data.sum(axis=2).sum(axis=1)
    sys = (np.abs(dataset_N_import.npred_background() - dataset_N_import.background)*map_mask).sum(axis=2).sum(axis=1)
    stat = np.sqrt(bg)
    signal = dataset_asimov_testing.npred_signal().data.sum(axis=2).sum(axis=1)

    '''print(spatial_model_type)
    print("With amplitude:", amplitude)
    print("BG counts:", bg)
    print("Signal counts:",signal )
    print("Ratio:",signal/bg )
    print()
    print("sys counts:",sys )
    print("Ratio sys:",sys/bg )'''
    return stat , sys, bg, signal

```
