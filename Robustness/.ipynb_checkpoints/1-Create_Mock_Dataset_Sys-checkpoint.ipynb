```python

def compute_K_matrix(l_deg, sigma, ndim_spatial_nui, ndim_spectral_nui,geom_down ):
    helper_map = Map.from_geom(geom_down).slice_by_idx(dict(energy=slice(0, 1)))
    helper_map2 = helper_map.copy()
    ndim_spatial_nui_1D = int(np.sqrt(ndim_spatial_nui))
    corr_matrix_spatial = np.identity(ndim_spatial_nui)
    for b_0 in range(ndim_spatial_nui_1D):
        for l_0 in range(ndim_spatial_nui_1D):
            i = b_0 * ndim_spatial_nui_1D + l_0
            C = SkyCoord(
                helper_map.geom.pix_to_coord((l_0, b_0, 0))[0],
                helper_map.geom.pix_to_coord((l_0, b_0, 0))[1],
                frame=geom_down.frame,
            )
            helper_map.data[0, :, :] = C.separation(
                geom_down.to_image().get_coord().skycoord
            ).value
            helper_map2.data = np.zeros(ndim_spatial_nui_1D ** 2).reshape(
                helper_map2.geom.data_shape
            )
            helper_map2.data[0, :, :] = np.exp(
                -0.5 * helper_map.data[0, :, :] ** 2 / l_deg ** 2
            )
            corr_matrix_spatial[i, :] = helper_map2.data.flatten()

    corr_matrix_spectral = np.identity(ndim_spectral_nui)
    for e in range((ndim_spectral_nui)):
        corr_matrix_spectral[e, e] = sigma[e] ** 2
    return np.kron(corr_matrix_spectral, corr_matrix_spatial)
```

```python
len_nui_per_ebin = 16
def plot_nuisance_histo (recoved_datasets, elist, masked):
    fig, axs = plt.subplots(len(elist)//3 +1, 3, figsize = (15,(len(elist)//3 +1 )* 3))
                           #sharex= True,)

    for i, e in enumerate(elist):
        ax = axs.flatten()[i]
        index_nui = e - 6 ## Starts with index 6!
        N_pars_input = dataset_N_import_624_co.N_parameters.value *100
        N_pars_input= N_pars_input[i*len_nui_per_ebin : (i+1)* len_nui_per_ebin]
        N_pars_input = [n for n in N_pars_input if n!=0  ] 
        mu, sigma = mus[e], stds[e] # mean and standard deviation
        s = np.random.normal(mu, sigma, len(N_pars_input)*1)\
    


        count, bins, ignored = ax.hist(s, 25, density=False, alpha = 0., color = 'white',)
        ax.plot(bins,   max(count) *
                            np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
                      linewidth=2, color='r')
        ax.fill_between(x = bins, y1 =  1e-2, y2= max(count) *
                            np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
                       color = 'r', alpha = 0.2)
        count, bins, ignored = ax.hist(
                N_pars_input,
                bins=20,
                alpha=0.7,
                label=""#Input: \n$\mu$ = {:.2} \n$\sigma$ = {:.2}".format(    np.mean(N_pars_input), np.std(N_pars_input))
                ,color = 'limegreen'
            )
        
        
            
        for j, ds in enumerate(recoved_datasets):
            try:
                N_pars_e = recoved_datasets[j].N_parameters.value*100
                N_pars_e= N_pars_e[i*len_nui_per_ebin : (i+1)* len_nui_per_ebin]
                if masked :
                    N_pars_e = N_pars_e* mask
                N_pars_e = [n for n in N_pars_e if n!=0  ] 

                _, bins, _ = ax.hist(
                    N_pars_e,
                    bins=bins,
                    alpha=0.8,
                    label=""#Recoved: \n$\mu$ = {:.2} \n$\sigma$ = {:.2}".format(np.mean(N_pars_e), np.std(N_pars_e))
                    ,color = "skyblue"
                )
                ax.errorbar(x =np.mean(N_pars_input), xerr = np.std(N_pars_input),
                       y = 15,fmt = 'o', color = 'darkgreen' )
                ax.errorbar(x =np.mean(N_pars_e), xerr = np.std(N_pars_e),
                       y = 20,fmt = 'o', color = 'blue' )
            except:
                temp = -0
            ax.set_xlabel("Sysetmatic [% of bg]")
            #ax.set_title(f"Energy: {[e]:.2}")
            ylim = ax.get_ylim()
            ax.vlines(0 +tots[i]  * 1, ylim[0], ylim[1],color = 'red' )
            #ax.vlines(0 -tots[i]  * 1, ylim[0], ylim[1],color = 'red')
        ax.set_yscale('log')
    plt.tight_layout()
    return s
    
    
```
